### Understand the Datasets

This assignment uses three datasets from the City of Chicago Data Portal:

1. **Socioeconomic Indicators in Chicago**  
   - Six public health–related socioeconomic indicators + hardship index (2008–2012).  
   - [Dataset link](https://data.cityofchicago.org/Health-Human-Services/Census-Data-Selected-socioeconomic-indicators-in-C/kn9c-c2s2)

2. **Chicago Public Schools**  
   - School-level performance data for CPS Report Cards (2011–2012).  
   - [Dataset link](https://data.cityofchicago.org/Education/Chicago-Public-Schools-Progress-Report-Cards-2011-/9xs2-f89t)

3. **Chicago Crime Data**  
   - Reported crime incidents (2001–present, excluding last 7 days).  
   - [Dataset link](https://data.cityofchicago.org/Public-Safety/Crimes-2001-to-present/ijzp-q8t2)

# 📥 Load CSV Data into Database
We load three datasets into SQLite tables:  
1. **Chicago Census Data** → `Chicago_Census_Data`  
2. **Chicago Public Schools** → `Chicago_Public_Schools`  
3. **Chicago Crime Data** → `Chicago_Crime_Data`


```python
census_data_df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01")
census_data_df.to_sql("Chicago_Census_Data", con, if_exists='replace', index=False, method="multi")
```

```python
public_schools_df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01") 
public_schools_df.to_sql("Chicago_Public_Schools", con, if_exists='replace', index=False, method="multi", chunksize=200)
```

```python
crime_data_df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01")
crime_data_df.to_sql("Chicago_Crime_Data", con, if_exists='replace', index=False, method="multi")
```

# 🔍 SQL Queries
We now run different queries to explore the data:


```python
%sql SELECT COUNT(*) AS Total_Crimes FROM Chicago_Crime_Data;
```

```python
%sql SELECT COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, PER_CAPITA_INCOME\
FROM Chicago_Census_Data\
WHERE PER_CAPITA_INCOME < 11000;
```

```python
%sql \
SELECT CASE_NUMBER \
FROM Chicago_Crime_Data \
WHERE DESCRIPTION LIKE '%MINOR%' AND DESCRIPTION NOT LIKE '%CHILD%';
```

```python
%sql SELECT CASE_NUMBER, PRIMARY_TYPE, DESCRIPTION \
FROM Chicago_Crime_Data \
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%';
```

```python
%sql SELECT DISTINCT PRIMARY_TYPE\
     FROM Chicago_Crime_Data\
     WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```

```python
%sql SELECT ["Elementary, Middle, or High School"] AS School_Type,\
       AVG(SAFETY_SCORE) AS Average_Safety_Score\
FROM Chicago_Public_Schools\
GROUP BY ["Elementary, Middle, or High School"];
```

```python
%sql SELECT COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY\
FROM Chicago_Census_Data\
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC\
LIMIT 5;
```
